```python
# Import  neceessary libaries for data cleaning,loading and data visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading a csv file named hr_data using pandas
hr_data=pd.read_csv("HR-Employee-Attrition.csv")
```

# Cleaning The data

```python
# Checks the sum of null values across columns in hr_data
hr_data.isnull().sum()
```

```python
# Checks the sum of duplicates in hr_data
hr_data.duplicated().sum()
```

```python
# Removing unnessary columns
hr_data=hr_data.drop(columns=['EmployeeCount','RelationshipSatisfaction','StandardHours','EmployeeNumber','DailyRate','Education','EnvironmentSatisfaction',
                              'JobInvolvement','StockOptionLevel','YearsInCurrentRole', 'YearsSinceLastPromotion','Over18',
       'YearsWithCurrManager', 'PerformanceRating',
       'TrainingTimesLastYear', 'WorkLifeBalance','PercentSalaryHike','MonthlyRate','JobSatisfaction','Attrition'])
```

```python
# Renaming names of columns
hr_data.rename(columns={'BusinessTravel':'Business Travel','DistanceFromHome'
                        :'Distance From Home','NumCompaniesWorked':'Number of Companies Worked',
                        'EducationField':'Education Field','JobRole':'Job Role',
                        'MaritalStatus':'Marital Status','MonthlyIncome':'Monthly Income','YearsAtCompany':'Years At Company',
'TotalWorkingYears':'Total Working Years','OverTime':'Over Time','Over18':'Over Eighteen','JobLevel':'Job Level',                     },
               inplace=True)
```

# Data Visualization

```python
# Matplotlib and seaborn are two visualization tools used to show the  visual representation of data in python.
```

```python
#  Extract numeric columns from the DataFrame
numeric_cols = hr_data.select_dtypes(include=['int64', 'float64'])
# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True)
# Shows title
plt.title('Correlation Heatmap for Numeric Values')
plt.show()
```

```python
# Bar chart of overtime column.
#Assuming 'Overtime' is the column containing Overtime information
# You  need to replace it with the actual column name from your dataset
Overtime = hr_data['Over Time'].value_counts()
#Plooting The Bar Chart using matplotlib.pyplot
plt.figure(figsize=(8, 6))
# Shows title of bar chart
plt.title('Overtime')
Overtime.plot(kind='bar', color='orange')
plt.show()
```

```python
# Pie chart of Marital Status column.
#Assuming 'Marital Status' is the column containing marital status information
# You  need to replace it with the actual column name from your dataset
marital_status_count = hr_data['Marital Status'].value_counts()
# Plotting the bar chart using matplotlib.pyplot
plt.figure(figsize=(8, 6)) 
plt.pie(marital_status_count, labels=marital_status_count.index, autopct='%1.1f%%', startangle=90)
# Shows title of bar chart
plt.title('Marital Status Distribution')
# Displaying the plot
plt.show()
```
